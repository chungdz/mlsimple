```python
header = pd.read_csv('data/header.tsv', sep='\t')
df = pd.read_csv('data/sample.tsv', sep='\t', names=header.columns)
```

```python
id_feature = ["m:AdId", "m:OrderId", "m:CampaignId", "m:AdvertiserId", "m:ClientID", "m:TagId", "m:PublisherFullDomainHash", "m:PublisherId", "m:UserAgentNormalizedHash","m:DeviceOSHash"]
```

```python
df[id_feature].values[0]
```

```python
idxdicts = []
for idname in tqdm(id_feature):
    cdict = {}
    idx = 0
    for v in df[idname]:
        if v not in cdict:
            cdict[v] = idx
            idx += 1
    idxdicts.append(cdict)
```

```python
for index, idname in enumerate(id_feature):
    cdict = idxdicts[index]
    df[idname] = df[idname].map(cdict)
```

```python
id_indexes = np.arange(len(flist), len(flist + id_feature))
id_indexes
```
