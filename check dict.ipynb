```python
# order id
def getInfo(curi):
    curd = c['freq'][curi]
    curname = c['all_ids'][curi]
    cdf = pd.DataFrame([[k, v] for k, v in curd.items()], columns=["id name", 'frequency'])
    print(curname, cdf.shape)
    vc = cdf['frequency'].value_counts()
    print(vc[:20].reset_index().values.tolist())
```

```python
for i in range(len(c['all_ids'])):
    getInfo(i)
```

```python
def getIndexDict(curi, masked=10):
    curd = c['freq'][curi]
    curname = c['all_ids'][curi]
    print(curname, len(curd))
    newd = {'<UNK>': 0}
    newidx = 1

    for k, v in curd.items():
        if v > masked:
            newd[k] = newidx
            newidx += 1
        else:
            newd[k] = 0
    
    return newd
```

```python
ndlist = [getIndexDict(i) for i in range(len(c['freq']))]
```

```python
metadict = {
    "all_features": c["all_features"],
    "to_minus": c["to_minus"],
    "to_div": c["to_div"],
    "all_ids": c["all_ids"],
    "dicts": ndlist,
    "total_count": c["total_count"]
}
```
